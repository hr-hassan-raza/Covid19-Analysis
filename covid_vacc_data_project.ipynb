```python
# Dependencies and Setup
import pandas as pd
import os
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python

# Load .csv file from Resources folder
covid_data = os.path.join("owid-covid-data.csv")

# Read purchase data file and store into Pandas data frame
covid_df = pd.read_csv(covid_data, encoding="utf-8")
```

```python
covid_df.head()
```

```python
covid_data_NA = covid_df.loc[(covid_df["continent"] == "North America")]
```

```python
covid_data_NA.head()
```

```python
covid_data_NA.describe()
```

```python
most_deaths=covid_data_NA[covid_df['date']=='2020-10-03'].sort_values('total_deaths',ascending=False)[1:11]
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title('10 Countries in North America in which most deaths occured due to Covid-19')
sns.barplot('location','total_deaths',data=most_deaths);
```

```python
most_cases=covid_data_NA[covid_data_NA['date']=='2020-10-03'].sort_values('total_cases',ascending=False)[1:11]
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title('10 Countries in North America which most covid_cases occured')
sns.barplot('location','total_cases',data=most_cases);
```

```python
covid_df['date'] = pd.to_datetime(covid_df['date'], format='%Y-%m-%d')#conversion of date column from object to datetime
```

```python
covid_df['Year']=covid_df['date'].apply(lambda year: year.year)
covid_df['Month']=covid_df['date'].apply(lambda month: month.month)
covid_df['Day']=covid_df['date'].apply(lambda day: day.day)
```

```python
month_map = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'} # dictionary for mapping months name to actual months in numbers provided in data frame for visualization
```

```python

sorted_df = covid_df[covid_df['location']=='United States'].sort_values(['Year','Month','Day'])
```

```python
grouped_df = sorted_df.groupby(['Year','Month']).sum()
grouped_df.reset_index(inplace=True)
grouped_df['Month'] = grouped_df['Month'].map(month_map)
plt.figure(figsize=(14,6))
sns.lineplot(y='new_cases', x='Month', data=grouped_df[:-1],sort=False,marker='o');
# group.plot(x='Month',y='new_cases')
plt.xticks(rotation=45)
plt.title('Number of New Cases per month in United States')
plt.ylabel('Number of New Cases')
plt.xlabel("Month");
```

```python
continent_population = covid_data_NA[covid_data_NA['date']=='2020-10-03'].groupby(['location']).mean().sort_values('total_cases',ascending=False)['total_cases'].to_frame()
continent_population = continent_population.head(5)
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title('Mean of total cases in North America States')
plt.pie(continent_population.total_cases);
labels = list(continent_population.index)
plt.pie(continent_population.total_cases,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=45)

plt.show()
```
